```python
# Declaracion de variables
var_ent_cad_nombreEstudiante = ""
var_ent_float_not_PP = 0.0
var_ent_float_not_SP = 0.0
var_ent_float_not_TP = 0.0
var_proSal_float_notaDef = 0.0 # Variable que almacena 
```

```python
# Capturas de datos 
var_ent_cad_nombreEstudiante = input("Nombre Estudiante ; ")
var_ent_float_not_PP = float(input("Primer Parcial ;"))
var_ent_float_not_SP = float(input("Segundo Parcial ;"))
var_ent_float_not_TP = float(input("Tercer Parcial ;"))
```

Nombre Estuandiante : Juan Camilo Lopez

```python
# Procesos
var_ent_float_nota_TP = (var_ent_float_not_PP+var_ent_float_not_SP+var_ent_float_not_TP)/3
```

```python
#  Proceso saber si el estudiante aprobo o reprobó
if(var_proSal_float_notaDef>=3.0):
  var_proSal_cad_concepto = "Aprobó - Te promueves"
else:
  var_proSal_cad_concepto = "Aprobo- Felicidades"
```

```python
#  Salidas
print("La nota definitiva de ",var_ent_cad_nombreEstudiante +" es : ",str(var_ent_float_nota_TP) + " Concepto : ",var_proSal_cad_concepto )
```
