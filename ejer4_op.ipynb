```python
# Declaracion de Variables
var_ent_float_velocidadIncial = 0.0
var_ent_float_velocidadFinal = 0.0
var_ent_float_aceleracion = 0.0
var_proSal_float_Tiempo = 0.0 # Variable que almacena 
```

```python
# Captura de datos
var_ent_float_velocidadIncial = float(input("velocida Incial ;"))
var_ent_float_velocidadFinal = float(input("velocidad Final ;"))
var_ent_float_aceleracion = float(input("aceleracion ;"))
```

Velocidad Incial; 12

Velocidad Final; 40

Aceleracion; 2


```python
# Procesos
var_proSal_float_Tiempo = (var_ent_float_velocidadFinal - var_ent_float_velocidadIncial)/var_ent_float_aceleracion
```

```python
# Salidas
print(var_proSal_float_Tiempo)
```
