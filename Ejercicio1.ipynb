Elaborar un algoritmo que solicite el numero de respuestas correctas, incorresta y blanco, correspondientes a postulaciones, y muestre su puntaje final considerando, que por cada respuesta correcta tendra 4 puntos, respuesta incorrectas tendra -1 y respuesta en blanco tendra 0.

```python
# Declaracion de variables
var_ent_int_RespuestasBuenas = 4
var_ent_int_RespuestasIncorrectas = -1
var_ent_int_RespuestasBlanco = 0
var_proSal_int_PuntajeFinal = 0.0 # Variable que almacena
```

```python
# Captura de datos
var_ent_int_RespuestasBuenas= float(input("Respuestas Buenas ;"))
var_ent_int_RespuestasIncorrectas = float(input("Respuestas Incorrectas ;"))
var_ent_int_RespuestasBlanco = float(input("Respuestas Blanco ;"))
```

```python
# Procesos
var_proSal_int_PuntajeFinal = (var_ent_int_RespuestasBuenas*4+var_ent_int_RespuestasIncorrectas*-1+var_ent_int_RespuestasBlanco*0)
```

```python
# Salidas
print(var_proSal_int_PuntajeFinal)
```
