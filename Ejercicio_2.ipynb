```python
# Desarrolo de la funcion 

def Ult_Dig(Num):
    UltDig = 0 
    UltDig = Num%100
    return(UltDig)

# Llamado de la funcion 

valor = int(input("Digite Numero : "))
print("Sus ultimos dos Digito son :", Ult_Dig(valor))
```
