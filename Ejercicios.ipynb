```python
#Desarrollo de la funcion 

def Ult_Dig(Num):
    UltDig = 0
    UltDig = Num%10
    return(UltDig)
#Llamado de la funcion

valor = int(input("Digite Numero : "))
print("El ultimo Digito es : ", Ult_Dig(valor))
```
